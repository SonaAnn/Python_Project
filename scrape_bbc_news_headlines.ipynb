```python
import requests
from bs4 import BeautifulSoup
import csv

# Function to scrape and display top news headlines
def scrape_bbc_news_headlines():
    url = 'https://www.bbc.com/news'

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        headline_elements = soup.find_all(class_='gs-c-promo-heading')
        
        # Print the headlines
        print("BBC News Headlines:")
        for idx, element in enumerate(headline_elements, start=1):
            headline_text = element.get_text(strip=True)
            print(f"{idx}. {headline_text}")
        
        # Create or open a CSV file to write headlines
        with open('bbc_news_headlines_1.csv', 'w', newline='', encoding='utf-8') as csvfile:
            # Create a CSV writer
            csv_writer = csv.writer(csvfile)
            
            # Write a header row to the CSV file
            csv_writer.writerow(["Index", "Headline"])

            # Extract and save the headlines to the CSV file
            for idx, element in enumerate(headline_elements, start=1):
                headline_text = element.get_text(strip=True)
                csv_writer.writerow([idx, headline_text])
                
        print("Headlines written to 'bbc_news_headlines_1.csv'")
    else:
        print("Failed to retrieve the web page. Status code:", response.status_code)

# Call the function to scrape and print headlines, and then create the CSV file
scrape_bbc_news_headlines()
```
